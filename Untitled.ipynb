```python
c = sort[['IN_CERTIFICADO','NU_IDADE']]
c = c[c['NU_IDADE']<=60]
c = c[c['IN_CERTIFICADO']==1].groupby('NU_IDADE')['IN_CERTIFICADO'].count()
c.plot()
plt.show()
```

```python
T = []
for i in c:
    T.append(i)
T
```

```python
maior_idade = sort[['NU_IDADE', 'IN_CERTIFICADO']]
maior_idade = maior_idade[maior_idade['NU_IDADE']>=18]
maior_idade = maior_idade[maior_idade['NU_IDADE']<=60]
maior_idade = maior_idade.groupby('NU_IDADE')['IN_CERTIFICADO'].count()
maior_idade
```

```python
i = 0
for j in maior_idade:
    T[i] = (i+18, T[i]/j)
    i+=1
T
```

```python
plt.scatter(*zip(*T))
plt.show()
```
